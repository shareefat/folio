```python
# break statement


x=0
while x<10:
  print("x is less than 10")
  if x==3:
    print("x is equal to 3")
    break
  x=x+1
```

```python
#continue statement
x=0
while x<10:
  print("x is lessthan 3")
  if x==3:
    print("x is equal to 3")
  continue
  break
  x=x+1

```

write a code to find the factorial using while loop

```python
n=int(input("enterthe number :"))
factorial=1
while n>1:
  factorial=factorial*n
  n=n-1
print(factorial)
```

```python
def facto(n):
  factorial=1
  for i in range(1,n+1):
    factorial=factorial*i
  return factorial
```

```python
facto(3)
```
