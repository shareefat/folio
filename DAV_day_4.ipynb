```python
import pandas as pd
temp_df=pd.DataFrame({
    "city":["mumbai","delhi","bangalore","kerala"],
    "temperature":[32,45,30,40]
})
temp_df
```

```python
import pandas as pd
windspeed_df=pd.DataFrame({
    "city":["mumbai","delhi","bangalore","karnataka"],
    "windspeed":[7,12,9,7]
})
windspeed_df
```

```python
df=pd.merge(temp_df,windspeed_df,on="city")
df
```

```python
df=pd.merge(temp_df,windspeed_df,on="city",how="left")  #to get the every in left
df
```

```python
df=pd.merge(temp_df,windspeed_df,on="city",how="right") #to get every value in right
df
```

```python
df=pd.merge(temp_df,windspeed_df,on="city",how="outer") # to get the full including left and right
df
```

```python
df=pd.merge(temp_df,windspeed_df,on="city",how="outer",indicator=True) #to indicate the values in the data frame
df
```

#cross tab

```python
df=pd.read_csv("survey.csv")
df
```

```python
pd.crosstab(df['Nationality'],df['Sex'])
```

```python
pd.crosstab(df['Nationality'],df['Handedness'],margins=True)
```

```python
pd.crosstab(df['Sex'],df["Handedness"],margins=True)
```

```python
pd.crosstab(df['Sex'],df["Handedness"],margins=True,normalize="index")
```

```python
pd.crosstab(df['Sex'],df["Handedness"],margins=True,normalize="columns")
```

```python
pd.crosstab(df['Sex'],df["Handedness"],margins=True,normalize="all")
```

#when we give values e need to give the aggrfun

```python
pd.crosstab(df['Nationality'],df["Handedness"],values=df["Age"],aggfunc="mean")
```
