How many players are playing in the NBA league

```python
import pandas as pd
nba=pd.read_csv('nba.csv')
nba
```

 Find all the players playing who belongs to Texas college in NBA league

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
nba[nba['College']=='Texas']
```

Find all the players in NBA league aged between (25 – 30) both inclusive

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
nba[(nba['Age']>=25) & (nba['Age']<=30)]
```

 Find the number of players playing for Philadelphia 76ers

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
nba[nba['Team']=='Philadelphia 76ers']
```

Find all the players in NBA league aged between (25 – 30) both inclusive and plays as Pointguard

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
nba[(nba['Age']>=25) & (nba['Age']<=30) & (nba['Position']=='PG')]
```
