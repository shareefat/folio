# CONCAT

```python
temp_df=pd.DataFrame({
    'city':['mumbai','delhi','bangalore'],
    'temperature':[32,45,30]
})
temp_df
```

```python
windspeed_df=pd.DataFrame({
    'city':['delhi','mumbai'],
    'windspeed':[7,12]
},index=[1,0])
windspeed_df
```

```python
pd.concat([temp_df,windspeed_df],axis=1)
```

# MERGE

```python
temp_df=pd.DataFrame({
    'city':['mumbai','delhi','bangalore'],
    'temperature':[32,45,30]
})
temp_df
temp_df
```

```python
windspeed_df=pd.DataFrame({
    'city':['delhi','mumbai'],
    'windspeed':[7,12]
})
windspeed_df
```

```python
df=pd.merge(temp_df,windspeed_df,on='city',how='outer')
df
```

```python
df=pd.merge(temp_df,windspeed_df,on='city',how='inner')
df
```
