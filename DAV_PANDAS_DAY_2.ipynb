#q)age <25 cant buy insurance >25 can buy insurance

```python
import numpy as np
age1= np.arange(15,25)
age2 = np.arange(25,35)
not_buy = np.zeros(10)
buy = np.ones(10)
```

```python
import pandas as pd
df1=pd.DataFrame(zip(age1,not_buy),columns=["AGE","BUY"])  #zip is used to align properly

df1
```

```python
df2=pd.DataFrame(zip(age2,buy),columns=["AGE","BUY"])
df2
```

```python
df1=pd.DataFrame(zip(age1,not_buy),columns=["AGE","BUY"])
df2=pd.DataFrame(zip(age2,buy),columns=["AGE","BUY"])
df=pd.concat([df1,df2],ignore_index=True) #normaly the index number will be like 0-9 after that it will be 0 again by using ignore_index=TRUE it dont be like that
df
df
```
