#series

```python
import pandas as pd
s=pd.Series (["Winter","Manson","Rain","Sunny","Snow"],name="WEATHER")
print(s)
```

```python
city=pd.Series(["Mumbai","Delhi","Goa","Kerala","Kashmir"],name="CITY")
c=pd.concat([s,city],axis=1)        #concat is used to combain the values into one
print(c)
```
